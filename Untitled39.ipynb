```python
from google import genai
from google.colab import files
from PIL import Image

API_KEY = "xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"

client = genai.Client(api_key=API_KEY)

uploaded = files.upload()

image_path = next(iter(uploaded))

image = Image.open(image_path)

prompt = """
You are a creative art assistant.

Analyze this artwork and generate:

1. A Catchy Artwork Title
2. A Detailed Artwork Description
3. An Instagram Caption with relevant hashtags
4. A Short Inspirational Story (100-150 words)
5. Suggested Color Palette
6. Art Style Identification
7. Tips to Improve the Artwork
"""

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=[image, prompt]
)

print("\n" + "="*60)
print("🎨 AI ART ANALYSIS")
print("="*60)
print(response.text)
```
